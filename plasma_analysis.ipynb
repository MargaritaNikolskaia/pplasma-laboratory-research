```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_normalized_intensity_profile(image_path):
    # Загрузка изображения в градациях серого
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Применение пороговой обработки для создания маски (поиск трубки)
    _, mask = cv2.threshold(image, 10, 255, cv2.THRESH_BINARY)

    # Удаление черных областей
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        tube_mask = np.zeros_like(image)
        cv2.drawContours(tube_mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

        masked_image = cv2.bitwise_and(image, image, mask=tube_mask)

        # Вывод маски
        plt.figure(figsize=(6, 6))
        plt.imshow(tube_mask, cmap='gray')
        plt.title('Маска трубки')
        plt.axis('off')
        plt.show()

        intensity_profile = np.mean(masked_image, axis=0)

        B_max = np.max(intensity_profile) if np.max(intensity_profile) != 0 else 1
        normalized_intensity_profile = intensity_profile / B_max
    else:
        normalized_intensity_profile = np.zeros(image.shape[1])

    return normalized_intensity_profile



image_path = 'air_1_torr.JPG'  # Убедитесь, что файл изображения доступен в рабочем каталоге
normalized_intensity_profile = calculate_normalized_intensity_profile(image_path)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calculate_normalized_intensity_profile1(image_path):
    # Загрузка изображения в градациях серого
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Применение пороговой обработки для создания маски (поиск трубки)
    # _, mask = cv2.threshold(image, 10, 255, cv2.THRESH_BINARY)


    mask = np.zeros_like(image)

    # Устанавливаем порог для каждого столбца
    for col in range(image.shape[1]):
        # Определяем пороговое значение для текущего столбца
        if 345 < col < 348 or 663 < col < 666 or 2500 < col < 2515 or 2391 < col < 2394:
            threshold_value = 0.3*(np.max(image[:, col]) - np.mean(image[:, col]))
        elif 2365 < col < 2368:
            threshold_value = 0.26*(np.max(image[:, col]) - np.mean(image[:, col]))
        elif 2600 < col < 2755:
            threshold_value = 2.5*np.mean(image[:, col])
        else :
            threshold_value = 0.34*(np.max(image[:, col]) - np.mean(image[:, col])) 
        
        _, temp_mask = cv2.threshold(image[:, col], threshold_value, 255, cv2.THRESH_BINARY)
        
        # Обновляем маску, объединяя текущую маску с основной
        mask[:, col] = np.maximum(mask[:, col], temp_mask[:, 0])


    # Удаление черных областей
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        tube_mask = np.zeros_like(image)
        cv2.drawContours(tube_mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

        masked_image = cv2.bitwise_and(image, image, mask=tube_mask)

        # Вывод маски
        plt.figure(figsize=(6, 6))
        plt.imshow(tube_mask, cmap='gray')
        plt.title('Маска трубки')
        plt.axis('off')
        plt.show()

        intensity_profile = np.mean(masked_image, axis=0)

        B_max = np.max(intensity_profile) if np.max(intensity_profile) != 0 else 1
        normalized_intensity_profile1 = intensity_profile / B_max
    else:
        normalized_intensity_profile1 = np.zeros(image.shape[1])

    return normalized_intensity_profile1

def calculate_normalized_intensity_profile2(image_path):
    # Загрузка изображения в градациях серого
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Применение пороговой обработки для создания маски (поиск трубки)
    # _, mask = cv2.threshold(image, 10, 255, cv2.THRESH_BINARY)


    mask = np.zeros_like(image)

    # Устанавливаем порог для каждого столбца
    for col in range(image.shape[1]):
        # Определяем пороговое значение для текущего столбца
        if 345 < col < 348 or 663 < col < 666 or 2500 < col < 2515 or 2391 < col < 2394:
            threshold_value = 0.3*(np.max(image[:, col]) - np.mean(image[:, col]))
        elif 2365 < col < 2368:
            threshold_value = 0.26*(np.max(image[:, col]) - np.mean(image[:, col]))
        else :
            threshold_value = 0.34*(np.max(image[:, col]) - np.mean(image[:, col])) 
        
        _, temp_mask = cv2.threshold(image[:, col], threshold_value, 255, cv2.THRESH_BINARY)
        
        # Обновляем маску, объединяя текущую маску с основной
        mask[:, col] = np.maximum(mask[:, col], temp_mask[:, 0])


    # Удаление черных областей
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if contours:
        largest_contour = max(contours, key=cv2.contourArea)
        tube_mask = np.zeros_like(image)
        cv2.drawContours(tube_mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

        masked_image = cv2.bitwise_and(image, image, mask=tube_mask)

        # Вывод маски
        plt.figure(figsize=(6, 6))
        plt.imshow(tube_mask, cmap='gray')
        plt.title('Маска трубки')
        plt.axis('off')
        plt.show()

        intensity_profile = np.mean(masked_image, axis=0)

        B_max = np.max(intensity_profile) if np.max(intensity_profile) != 0 else 1
        normalized_intensity_profile2 = intensity_profile / B_max
    else:
        normalized_intensity_profile2 = np.zeros(image.shape[1])

    return normalized_intensity_profile2

# Пример использования для одного изображения
image_path = 'air_1_torr.JPG'  # Убедитесь, что файл изображения доступен в рабочем каталоге
normalized_intensity_profile1 = calculate_normalized_intensity_profile1(image_path)
normalized_intensity_profile2 = calculate_normalized_intensity_profile2(image_path)

# Построение графика
plt.plot(normalized_intensity_profile1, color='red')
plt.plot(normalized_intensity_profile2)
plt.xlabel('Z (положение вдоль оси трубки)')
plt.ylabel('Интенсивность (B / Bmax)')
plt.title('Нормированный профиль интенсивности вдоль оси Z')
plt.show()
plt.savefig('intensity_profile_air_1_torr.png')
```
